```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file, explicitly defining the header
data = pd.read_csv("vertex_data_1.csv", names=["Time", "Position_X", "Position_Y", "Position_Z", "Velocity_X", "Velocity_Y", "Velocity_Z", "Force_X", "Force_Y", "Force_Z", "BendingForce_X", "BendingForce_Y", "BendingForce_Z", "DampingForce_X", "DampingForce_Y", "DampingForce_Z"], header=0)

# Plot Velocity
plt.figure()
plt.plot(data['Time'], data['Velocity_X'], label="Velocity X")
plt.plot(data['Time'], data['Velocity_Y'], label="Velocity Y")
plt.plot(data['Time'], data['Velocity_Z'], label="Velocity Z")
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity of Vertex 0 over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot Force
plt.figure()
plt.plot(data['Time'], data['Force_X'], label="Force X")
plt.plot(data['Time'], data['Force_Y'], label="Force Y")
plt.plot(data['Time'], data['Force_Z'], label="Force Z")
plt.xlabel('Time (s)')
plt.ylabel('Force (N)')
plt.title('Force on Vertex 0 over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot Force
plt.figure()
plt.plot(data['Time'], data['BendingForce_X'], label="BendingForce X")
plt.plot(data['Time'], data['BendingForce_Y'], label="BendingForce Y")
plt.plot(data['Time'], data['BendingForce_Z'], label="BendingForce Z")
plt.xlabel('Time (s)')
plt.ylabel('Bending Force (N)')
plt.title('Bending Force on Vertex 0 over Time')
plt.legend()
plt.grid(True)
plt.show()


# Plot Force
plt.figure()
plt.plot(data['Time'], data['DampingForce_X'], label="DampingForce X")
plt.plot(data['Time'], data['DampingForce_Y'], label="DampingForce Y")
plt.plot(data['Time'], data['DampingForce_Z'], label="DampingForce Z")
plt.xlabel('Time (s)')
plt.ylabel('Damping Force (N)')
plt.title('Damping Force on Vertex 0 over Time')
plt.legend()
plt.grid(True)
plt.show()


# Plot Position
plt.figure()
plt.plot(data['Time'], data['Position_X'], label="Position X")
plt.plot(data['Time'], data['Position_Y'], label="Position Y")
plt.plot(data['Time'], data['Position_Z'], label="Position Z")
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position of Vertex 0 over Time')
plt.legend()
plt.grid(True)
plt.show()
```
